```python
import pandas as pd
from scipy.stats import spearmanr
import pyreadstat

# Load the data
df, meta = pyreadstat.read_sav('C:\\Users\\yassi\\Downloads\\TEST_1.sav')

# Get unique grades
grades = df['Grade'].unique()

# Create a new DataFrame to store the results
results_df = pd.DataFrame(columns=['Grade', 'Correlation', 'P_value'])

# For each grade, calculate the correlation
for grade in grades:
    df_grade = df[df['Grade'] == grade].copy()  # Take a subset of the DataFrame for this grade
    df_grade.dropna(subset=['Median'], inplace=True)  # Drop rows where 'Median' is NaN
    df_grade['Rank_within_grade'] = range(1, len(df_grade) + 1)  # Create a ranking variable within the grade

    if len(df_grade['Median'].unique()) > 1:  # Check if 'Median' has more than one unique value
        corr, p_val = spearmanr(df_grade['Rank_within_grade'], df_grade['Median'])
    else:
        corr, p_val = float('nan'), float('nan')  # Set correlation and p-value to nan if 'Median' has no variance

    # Append the results to the results DataFrame
    results_df = results_df.append({'Grade': grade, 'Correlation': corr, 'P_value': p_val}, ignore_index=True)


# Display the results DataFrame
print(results_df.drop([0]))
```
